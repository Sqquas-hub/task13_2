```python
import pandas as pd
from keybert import KeyBERT  # или yake

def find_keywords(file_path):
    # 1. Загружаем данные
    df = pd.read_csv(file_path)

    # 2. Инициализируем модель
    kw_model = KeyBERT()  # или kw_extractor = yake.KeywordExtractor()

    # 3. Создаем пустой список для результатов
    keywords_list = []

    # 4. Для каждого отзыва извлекаем ключевые слова
    for text in df['normalized_text']:
        if pd.notna(text) and text.strip():
            keywords = kw_model.extract_keywords(
                text,
                keyphrase_ngram_range=(1, 2),  # можно искать фразы из 1-2 слов
                stop_words='russian',           # удаляем стоп-слова
                top_n=5                         # берем топ-5 ключевых слов
            )
        else:
            keywords = []
        keywords_list.append(keywords)

    # 5. Добавляем новый столбец
    df['key_words'] = keywords_list

    # 6. Возвращаем обновленный DataFrame
    return df

result = find_keywords('cafe_reviews.csv')
print(result[['normalized_text', 'key_words']].head())
```
